```python
import threading
import time

a = 0
b = 0
stop_threads = False

def increase_a():
    global a
    while not stop_threads:
        time.sleep(1)
        a += 1
        print(f"a: {a}")

def increase_b():
    global b
    while not stop_threads:
        time.sleep(2)
        b += 1
        print(f"b: {b}")

t1 = threading.Thread(target=increase_a)
t2 = threading.Thread(target=increase_b)

t1.start()
t2.start()

try:
    while True:
        pass
except KeyboardInterrupt:
    stop_threads = True
    t1.join()
    t2.join()
    print("Threads stopped.")
```
